### 2.1.1 Prompts for LLaMa

```python
import ollama
from IPython.display import Markdown, display

def llama_chat(prompt):
    client = ollama.Client(host='http://localhost:11434')
    response = client.chat(model='llama3', messages=[{"role": "user", "content": prompt}])
    return response["message"]["content"]
    
llama_output = llama_chat('Suggest a 3-day itinerary in Tokyo for a solo traveler who loves Manga.')
display(Markdown(llama_output))
```

## 3.3. Developing a Frontend 

The frontend is implemented as a Streamlit web app (`travel_assistant.py`), providing an interactive interface for the travel assistant.

Streamlit is an open-source Python library that makes it easy to create and share custom web apps for machine learning and data science

To run Streamlit frontend:
Steps : 
- In the terminal 'pip install streamlit requests / conda install streamlit requests '
- run : streamlit run travel_assistant_.py from the folder.

It includes:

- **Form**: Collects name, destination, dates, budget, interests, group size and places to avoid.
- **Dialogue Display**: Shows the user inputs and Assistant responses in a conversational (chat) format.
- **Itinerary Output**: Displays the LLaMA 3-generated itinerary.

**Implementation**:
- File: `travel_assistant_.py`
- Features: Form inputs, error handling, styled output using Streamlit components.
- Queries LLaMA 3 via Ollama’s API.
- Run: `streamlit run travel_assistant_.py` 


```python
#Code for travel_assistant_.py

import streamlit as st
import requests
import json

st.set_page_config(page_title="Travel Assistant", page_icon="🌍✈️")
st.sidebar.title("🌐 Trip Info")
st.title("🤖 AI Travel Chat Assistant")

# Initializing session state
if "step" not in st.session_state:
    st.session_state.step = 0
if "user_info" not in st.session_state:
    st.session_state.user_info = {
        "name": "",
        "destination": "",
        "days": "",
        "user_count": "",
        "interests": "",
        "budget": "",
        "exclusions": ""
    }
if "final_response" not in st.session_state:
    st.session_state.final_response = ""
if "messages" not in st.session_state:
    st.session_state.messages = [
        {
            "role": "assistant",
            "content": (
                "Hi! I’m your travel assistant\n\n"
                "Let’s plan your trip!\n"
            )
        }
    ]

#previous messages
for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

#user inout
user_input = st.chat_input("Say something to your travel assistant...")

if user_input:
    st.session_state.messages.append({"role": "user", "content": user_input})
    with st.chat_message("user"):
        st.markdown(user_input)

    info = st.session_state.user_info
    step = st.session_state.step

    # Step-by-step dialogue
    if step == 0:
        assistant_reply = "What's your name, destination and number of days? (e.g., Priya, Cairns, 4)"
        st.session_state.step += 1

    elif step == 1:
        try:
            parts = [x.strip() for x in user_input.split(",")]
            info["name"], info["destination"], info["days"] = parts[0], parts[1], parts[2]
            assistant_reply = "How many people are traveling?"
            st.session_state.step += 1
        except:
            assistant_reply = "Oops! Please use the format: Name, Destination, Days"

    elif step == 2:
        info["user_count"] = user_input
        assistant_reply = "Great! What are your travel interests? (e.g., beaches, food, nature, shopping)"
        st.session_state.step += 1

    elif step == 3:
        info["interests"] = user_input
        assistant_reply = "Awesome! What is your budget for this trip?"
        st.session_state.step += 1

    elif step == 4:
        info["budget"] = user_input
        assistant_reply = "Are there any places you'd like to avoid?"
        st.session_state.step += 1

    elif step == 5:
        info["exclusions"] = user_input

        # Generate the final prompt
        prompt = (
            f"Plan a {info['days']}-day travel itinerary for {info['name']} who wants to visit {info['destination']} "
            f"with {info['user_count']} people. Their interests are: {info['interests']}, within a budget of {info['budget']}. "
        )
        if info["exclusions"].strip():
            prompt += f"Exclude these places: {info['exclusions']}."

        # Send to LLaMA 3 (local)
        with st.chat_message("assistant"):
            with st.spinner("Planning your trip..."):
                try:
                    response = requests.post("http://localhost:11434/api/generate", json={
                        "model": "llama3",
                        "prompt": prompt
                    }, stream=True)

                    full_reply = ""
                    for line in response.iter_lines():
                        if line:
                            data = json.loads(line.decode("utf-8"))
                            full_reply += data.get("response", "")

                    assistant_reply = full_reply.strip()
                    st.session_state.final_response = assistant_reply
                    st.session_state.step += 1

                except Exception as e:
                    assistant_reply = f"Something went wrong: {e}"

    elif step == 6:
        if "yes" in user_input.lower():
            st.session_state.user_info = {
                "name": "",
                "destination": "",
                "days": "",
                "user_count": "",
                "interests": "",
                "budget": "",
                "exclusions": ""
            }
            st.session_state.final_response = ""
             
            assistant_reply = "Awesome! What's your name, destination and number of days?"
        elif "no" in user_input.lower():
            assistant_reply = "Great! Enjoy your trip. Let me know whenever you want to plan another one. Safe travels ✈️"
        else:
            assistant_reply = "Would you like to plan another trip? (yes or no)"

    # Append assistant reply to chat
    st.session_state.messages.append({"role": "assistant", "content": assistant_reply})
    with st.chat_message("assistant"):
        st.markdown(assistant_reply)
```

## 4. Prompt & Model Comparison

To see how prompt specificity influences the quality of responses generated by the LLM, we will be testing on two prompts using the same model (`llama3`) under identical conditions.

- **Prompt A**: A generic travel request
- **Prompt B**: A detailed and personalized travel plan for a birthday trip

The responses below demonstrate how increasing prompt clarity and personalization improves the usefulness and relevance of the output.


### 4.1 Comparing prompts

```python
short_prompt = (
    "Plan a 3-day trip to Brisbane."
)

llama_output = llama_chat(short_prompt)
display(Markdown("### Simple Prompt"))
display(Markdown(llama_output))
```

```python
clear_prompt = (
    "Plan a 3-day trip to Brisbane with my sister for her birthday. She loves beaches, arts, music, scenic viewpoint. She also enjoys peaceful walks, bike rides, and vegan food "
)

llama_output = llama_chat(clear_prompt)
display(Markdown("### Elaborated customized prompt"))
display(Markdown(llama_output))
```

The detailed prompt gave a better elaborated personalized output.

### 4.2 Comparing models

```python
!ollama pull gemma3:1b
```

```python
from IPython.display import Markdown, display
comparing_prompt = (
    "Plan a 3-day trip to Brisbane with my sister for her birthday. "
    "She loves beaches, arts, music, scenic viewpoints. She also enjoys peaceful walks, bike rides, and vegan food."
)

def gemmachat(prompt):
    client = ollama.Client(host='http://localhost:11434')
    response = client.chat(model="gemma3:1b", messages=[{"role": "user", "content": prompt}])
    return response["message"]["content"]
    
response = gemmachat(comparing_prompt)
display(Markdown(response))
```


### 4.2.1 Observations:

- LLaMA 3 provided detailed, well-structured responses suitable for a personal trip planner.

- Mistral was faster and offered creative suggestions but sometimes less consistent in structure.

- Gemma 3B was lightweight, efficient, and performed well, delivering friendly structured and empathetic plans.

All models responded well to tailored prompts, with minor variations in tone, length, and clarity.
